#帳號密碼判斷

```python
try:
  ID="tony"
  PWD="123"
  while True:
    x=input("請輸入帳號:")
    y=input("請輸入密碼:")
    if(x==ID and y==PWD):
      print("歡迎光臨!!!")
      break;
    else:
      print("帳號或密碼不符合。")
except Exception as e:
  print(e)
finally:
  print("end")
```

#猜數字

```python
import random
number=random.randint(1,16)
try:
  while True:
    print(number)
    guess=input("猜數字(1-15)")
    if(number==guess):
      print("猜對了")
      inputt=input("是否繼續")
      if(inputt=="Y"):
        number=random.randint(1.16)
      else:
        break;
    else:
      print("猜錯了")
except Exception as e:
  print(e)
finally:
  print("end")
```
