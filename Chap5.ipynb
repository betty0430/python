#函式
-獨立程式碼 可重複

-可以0-多個輸入參數

-輸入參數可有預設值

```python
#1.
import math
try:
  number=eval(input("請輸入圓半徑"))
  all=number*number*math.pi
  print("圓面積為:",all)
except Exception as e:
  print(e)
finally:
  print()
```

```python
#2
import math
try:
  def circle_area(input):
    #半徑*半徑*π
    all=input*input*math.pi
    print("圓面積為:",all)
  circle_area(5)
  circle_area(10)
except Exception as e:
  print(e)
finally:
  print()
```
