## if,else if

```python
#成績判別
try:
  while True:
    a=int(input("請輸入成績"))
    if a>=95:
      print("優等")
    elif a>=80:
      print("甲等")
    elif a>=60:
      print("乙等")
    elif a>=40:
      print("丙等")
    else:
      print("丁等")
except Exception as e:
  print(e)
finally:
  print("byby")
```

```python
a=int(input("請輸入a值"))#輸入一數值並且轉成數字
b=int(input("請輸入b值"))
#判斷並顯示<>=
if(a==b):
  print("兩數相等")
elif(a<b):
  print("a<b")
else:
  print("a>b")
```

### **for**



```python
try:
  a=int(input("請輸入數字1:"))
  b=int(input("請輸入數字2:"))
  if a>b:
    c=b
    b=a
    a=c
  
  for i in range(a,b,1):
    print(i)
except Exception as e:
  print(e)
finally:
  print("byby")
```

```python
try:
  #1.
  print("1.")
  for i in "123456786":
    print(i)
  #2.
  print("2.")
  for a in["a","b","c"]:
    print(a)
except Exception as e:
  print(e)
finally:
  print("byby")
```

```python
try:
  a=int(input("請輸入數字1:"))
  b=int(input("請輸入數字2:"))
  if a>b:
    c=b
    b=a
    a=c
  sum=0
  for d in range(a,b,1):
    sum+=d
  print(a,"+2+3+...+",b,"=",sum)#可以用，把字串和數字串連在一起
except Exception as e:
  print(e)
finally:
  print("byby")
```

```python
try:
  sum=0
  i=1
  while i < 11:
    sum+=i
    i+=1
  print(sum)
except Exception as e:
  print(e)
finally:
  print("byby")
```

```python
try:
  for chicken in range (0,36):
    rabbit=35-chicken
    if(2*chicken+4*rabbit)==94:
      print("雞有",chicken,"隻，兔子有",rabbit,"隻")
      break
    else:
      print("無解")
except Exception as e:
  print(e)
finally:
  print("byby")
```
