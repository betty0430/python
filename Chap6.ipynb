##檔案處裡
處理目錄與文件，以及檔案的複寫

```python
import os
try:
  print(os.getcwd()) #回傳目前的工作目錄
except Exception as e:
  print(e)
finally:
  print("")
```

```python
import os
try:
  #目前目錄上一個目錄下的檔案 ".."
  #目錄下的檔案 "."
  os.listdir(".")
except Exception as e:
  print(e)
finally:
  print("")
```

```python
import os
try:
  #新增檔案
  os.makedirs("123")
  os.listdir(".")
except Exception as e:
  print(e)
finally:
  print("")
```

```python
import os
try:
  #刪除檔案
  os.rmdir("123")
  os.listdir(".")
except Exception as e:
  print(e)
finally:
  print("")
```

##檔案處理

w = 可寫(預設建立檔案) 若檔案存在則會清除舊檔案

w+ = 可寫可讀(預設建立檔案) 若檔案存在則會清除舊檔案

r = 可讀

r+ = 可寫可讀 但寫入資料指向開頭

a = 可寫不能讀 寫入資料指向尾端

a+ = 可寫可讀 寫入資料指向尾端

b = 轉為2進位文件

ab+ = 2進位文件的讀寫 寫入資料指向尾端

```python
import os
try:
  #新增檔案
  a=open("a.txt","w") 
  a.write("This is a test1")
  a.write("This is a test2\n")
  a.write("This is a test3\n")
  a.close() #關閉檔案
except Exception as e:
  print(e)
finally:
  print("")
```

```python
#檔案是否存在
os.path.isfile("a.txt")
```

```python
import os
a=open("a.text","r")
#把檔案內的文字讀出
print(a.readlines())
a.close()
```

```python
#檔案是否存在
if os.path.isfile("a.txt"):
  a=open("a.txt","r")
  print(a.readlines())
  a.close()
else:
  print("檔案不存在")
```

```python
#會把輸入的檔案加入檔案的最後
a=open("a.text","a")
a.write("This is a test4\n")
a.close()
```

```python
#資料附加在檔案開頭
a=open("a.text","r+")
a.write("This is a test4\n")
a.close()
```
