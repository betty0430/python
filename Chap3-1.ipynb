Tuple (元組)
tuple(元組)可視為不可改變的串列，tuple跟list很像，但是我們不能新增、刪除或更新tuple的元素。tuple的好處有：

1.佔用較少空間

2.可當作dictionary(字典)的鍵值

3.可當作函式引數

```python
try:
  a=['1','2','3']
  print(a)
  b=['a','b','c']
  # 0  1 2
  b_tuple=tuple(b)
  print(b_tuple)
  print(b_tuple[2])
except Exception as e:
  print(e)
```

字典(dictionary - dict)為帶有鍵值(key)的串列(list)

```python
try:
  q={0:"a",1:"b",2:"c",3:"d",4:"e"}
  print(q)
  print(q[1])
  w={'fish':'魚','book':'書','egg':'蛋','phone':'手機'}
  print(w['fish'])
except Exception as e:
  print(e)
```
